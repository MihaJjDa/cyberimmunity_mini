# Название продукта / решения

Система управления пассажирским поездом (тягой, торможением, дверями, освещением, кондиционированием; автоведение по заданному расписанию)

# Концепция продукта

## Краткое описание назначения и применения продукта

Продукт предназначен для автономного управления движением пригородных поездов по заведомо составленным маршрутам по определенному расписанию с целью осуществления пассажирских перевозок. Процесс пассажирских перевозок подразумевает посадку и высадку пассажиров на остановочных пунктах маршрутов, а также обеспечение стандартизированных условий перевозки внутри вагонов (температура и освещение)

## Ценности продукта

Активами продукта являются
- Управляемый системой поезд
- Используемая управлемой системой поездом железнодорожная инфраструктура
- Находящиеся в поезде пассажиры
- Выполнение маршрута (?) - прохождение требуемых остановок не ранее соответствующему остановкам времени

*Последний пункт неочевиден - возник вопрос в том, как внести в активы не только технику безопасности, но и функциональное требование*

## Неприемлемые события или сценарии в отношении ценностей продукта

- Столкновение с другим поездом или элементом железнодорожной инфраструктуры
- Нанесение вреда жизни и здоровью пассажиру
- Пространственное или временное отклонение от выполнения маршрута при отсутствии объективных причин невозможности выполнения (?)

*Последний пункт спорный - не уверен, что это является неприемлемым событием, ведь даже управляемые человеком электрички могут не соответствовать маршруту без объективных причин (человеческий фактор), однако сделал вывод, что если есть актив - должен быть ущерб*

## Риски неприемлемых событий или сценариев в отношении ценностей продукта

- Нарушение управления поездом
- Выход из строя оборудования поезда 
- Некорректное принятие решений о прохождении маршрута (выход на маршрут, совершение посадки или высадки, завершение маршрута)
- Допущение или поддержание противоречащих безопасности жизни и здоровья пассажиров условий перевозки

# Цели безопасности

## Формулировки

1. Автономное управление поездом не может привести к столкновению с другими поездами или элементами железнодорожной структуры из-за движения
2. Автономное управление поездом не допускает пространственное или временное отклонение от заданного маршрута на основе принятия решений программным обеспечением системы
3. Автономное управление поездом не приводит к нанесению вреда жизни и здоровью пассажиров посредством движения поезда (ускорение и торможение) или посредством элементов поезда (открытие или закрытие дверей, небезопасное освещение, температурный режим)

## Предположения безопасности

1. Обеспечена пригодность использования оборудование анализа окружения поезда и обеспечена возможность приведения данного оборудования в пригодное состояние в ограниченное время (до 10 минут) во время прохождения маршрута
2. Остановочные пункты маршрутов исключают возможность нанесения вреда жизни и здоровью пассажиров при посадке и высадке при соблюдении правил посадки и высадки
3. Пассажиры соблюдают правила передвижения в поезде
